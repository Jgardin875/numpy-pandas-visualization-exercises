```python
x =np.array(range(0, 15))

y = np.sqrt(x)
plt.plot(x, y)
```

```python
x =np.array(range(0, 15))

y = x**3
plt.plot(x, y)
```

```python
x =np.array(range(0, 15))

y = 2**x
plt.plot(x, y) 
```

```python
x =np.array(range(0, 15))

y = 1/(x+1)
plt.plot(x, y) 
```

```python
x =np.array(range(0, 15))

plt.subplot(2,2,1)
y = np.sqrt(x)
plt.plot(x, y)

plt.subplot(2,2,2)
y = x**3
plt.plot(x, y)

plt.subplot(2,2,3)
y = 2**x
plt.plot(x, y) 


plt.subplot(2,2,4)
y = 1/(x+1)
plt.plot(x, y) 

```

```python
x =np.array(range(0, 3))

y1 = np.sqrt(x)
plt.plot(x, y1, 'b', label = '$sqrt(y)$')

y2 = x**3
plt.plot(x, y2, 'g', label = '$x^3$')

y3 = 2**x
plt.plot(x, y3, c='r', label = '$2^x$') 


y4 = 1/(x+1)
plt.plot(x, y4, c='m', label = '$1/(x+1)$') 

plt.title("Various Math Functions")
plt.xlabel('x-axis')
plt.ylabel('y-output')

plt.legend()
```
