```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.array (range (-10, 11, 1))
y = x**2 - x +2
plt.plot(x,y)
plt.text(-1,10, '(0,0)', fontsize = 10, color = 'b')
#arrowprops={}
#plt.savefit('figure1.png or jpeg or etc')
#%config InlineBackend.figure_format - 'svg'
#above line helps make it pretty in case you export it and need to zoom in or out on itlater
```

```python
x =np.array(range(0, 15))

y = np.sqrt(x)
plt.plot(x, y)
plt.title("Math Functions")
plt.xlabel('x-axis')
plt.ylabel('sqrt')
```

```python
x =np.array(range(0, 15))

y = x**3
plt.plot(x, y)
plt.title("Math Functions")
plt.xlabel('x-axis')
plt.ylabel('cubed')
#plt.ylim(0,100)
```

```python
x =np.array(range(0, 15))

y = 2**x
plt.plot(x, y) 
plt.title("Math Functions")
plt.xlabel('x-axis')
plt.ylabel('$2^x$')
```

```python
x =np.array(range(0, 15))

y = 1/(x+1)
plt.plot(x, y)
plt.title("Math Functions")
plt.xlabel('x-axis')
plt.ylabel('$y-output$')
```

```python
x =np.array(range(0, 15))

plt.suptitle('Various Math Functions')

plt.subplot(2,2,1)
y = np.sqrt(x)
plt.plot(x, y)
plt.xlabel('x-axis')
plt.ylabel('sqrt')

plt.subplot(2,2,2)
y = x**3
plt.plot(x, y)
plt.xlabel('x-axis')
plt.ylabel('cubed')

plt.subplot(2,2,3)
y = 2**x
plt.plot(x, y) 
plt.xlabel('x-axis')
plt.ylabel('$2^x$')

plt.subplot(2,2,4)
y = 1/(x+1)
plt.plot(x, y)
plt.xlabel('x-axis')
plt.ylabel('$2^x$')

plt.tight_layout()
```

```python
x =np.array(range(0, 3))

y1 = np.sqrt(x)
plt.plot(x, y1, 'b', label = '$sqrt(y)$')

y2 = x**3
plt.plot(x, y2, 'g', label = '$x^3$')

y3 = 2**x
plt.plot(x, y3, c='r', label = '$2^x$') 


y4 = 1/(x+1)
plt.plot(x, y4, c='m', label = '$1/(x+1)$') 

plt.title("Various Math Functions")
plt.xlabel('x-axis')
plt.ylabel('y-output')

plt.legend()
```
