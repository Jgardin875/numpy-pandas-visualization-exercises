```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
n = np.array(range(1, 4))

y1 =n + 1
plt.plot(n,y1, label = '$O(1)$')

y2 = np.log10(n)
plt.plot(n,y2, label = '$O(log n)$')


y3 = n
plt.plot(n,y3, label = '$O(n)$')

y4 = n*np.log10(n)
plt.plot(n,y4, label = '$O(n log n)$')

y5 = n**2
plt.plot(n,y5, label = '$O(n^2)$')

y6 = 2**n
plt.plot(n,y6, label = '$O(2^n)$')

#y7 = np.math.factorial(n)
#plt.plot(n,y7, label = '$O(n!)$')

y8 = n**n
plt.plot(n,y8, label = '$O(n^n)$')

plt.title('Big O Notation')
plt.xlabel('Elements')
plt.ylabel('Operations')
plt.legend()
```
