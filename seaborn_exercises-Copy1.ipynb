```python
#Using the chipotle dataset from the previous exercise, 
#create a bar chart that shows the 4 most popular items and the revenue produced by each.
```

```python
import env
```

```python
chp = pd.read_sql('SELECT * FROM orders', env.get_db_url('chipotle') )
chp.head()
```

```python
chp.item_price = chp.item_price.str.replace('$', '').str.strip().astype(dtype = 'float') #cleaning
```

```python
chp.groupby('item_name')[['quantity']].sum().sort_values('quantity', ascending = False).head(4)
```

```python
total_sold= chp.groupby('item_name')[['quantity']].sum().sort_values('quantity', ascending = False)
total_sold.head()
```

```python
avg_cost_per_item = chp.groupby('item_name')[['item_price']].mean()
avg_cost_per_item.head()
```

```python
rev = total_sold.merge(avg_cost_per_item, how = 'left', on = 'item_name')
```

```python
rev.head(4)
```

```python
rev['total_rev'] = (rev.quantity * rev.item_price)
rev.head(4)
```

```python
rev.info()
```

```python
rev_barplot = rev.sort_values('total_rev', ascending = False).head(4)
rev_barplot
```

```python
sns.barplot(data = rev_barplot, x = rev_barplot.index, y = 'total_rev')
```
